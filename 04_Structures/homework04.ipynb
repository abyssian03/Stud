#### Задача 4 (***) - смерть программиста

Разберите условие задачи и код ниже. Напишите в комментариях к коду, как вы понимаете принцип его работы и почему он решает данную задачу.

"Вася хочет позвать друзей на день рождения. Он знает, что некоторые его друзья дружат друг с другом, а некоторые - наоборот. Вика очень дружит с Машей и обязательно приведет ее с собой, если только не придет Марина. С другой стороны, если Маша придет без Вики, то Марина тоже придет. Марина точно придет, если придет Костя, а Костя придет, если не придет Петя. Петя придет, только если придут или Марина, или Маша. Боря придет в том и только том случае, если не придет Вика.

Программа принимает на вход одно из имен: "Вика", "Маша", "Марина", "Костя", "Петя" или "Боря" в зависимости от того, кого из них первым решил позвать Вася. На выходе программа выдает все возможные множества друзей Васи, которые могут прийти на его день рождения. Считается, что тот, кого Вася позвал первым, точно придет, а остальных Вася будет звать с учетом их предпочтений."

```python
names = ["Вика", "Маша", "Марина", "Костя", "Петя", "Боря"]
cond_lst = [[1, -3], [1, 3], [2, -1], [4], [-5], [-2, -3], [2], [3], [6], [-1]]
cons_lst = [2, -2, 3, 3, 4, -5, 5, 5, -1, 6]
res = []

first = input()
num = names.index(first) + 1

for i, cond in enumerate(cond_lst):
    res.append(set())
    if -num not in cond and -num != cons_lst[i]:
        res[i].add(num)
        for elem in cond:
            res[i].add(elem)
        res[i].add(cons_lst[i])

new_res = []
for elem in res:
    if elem != set() and elem not in new_res:
        new_res.append(elem)

print(new_res)

tpl_set = set()
for sett in new_res:
    new_sett = sett.copy()
    for sett1 in new_res:
        flag = False
        for num in sett1:
            if -num in new_sett:
                flag = True
        if flag == False:
            new_sett = new_sett | sett1

    tpl_set.add(tuple(new_sett))
print(tpl_set)

for tpl in tpl_set:
    for i, name in enumerate(names):
        if (i+1) in tpl:
            print(name, end=" ")
    print()
```
