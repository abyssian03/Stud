## Структуры данных в Python

Объекты в питоне, помимо типов, еще обладают некоторыми признаками, по которым их можно объединять в (пересекающиеся) группы. Таких признаков несколько:
- итерируемые
- Изменяемые / неизменяемые (hashable)
- вызываемые

Итерируемые объекты - такие, которые можно перебрать поэлементно. К ним относятся, например, строки.

Неизменяемые, они же хешируемые - такие, которые нельзя изменить по частям, а только перезаписать полностью. Это те, которые мы обсуждали все это время: числовые типа данных и строки. Хешируемыми они называются потому, что для того, чтобы отыскать такой объект в памяти, питон вычисляет т.н. hash-функцию и приписывает объекту определенный номер в таблице (хеш-таблице). Такой поиск осуществляется быстро, почти мгновенно, поэтому работа с неизменяемыми объектами экономит вычислительные ресурсы.

Вызываемые объекты - это функции. Их мы изучим позже.

Итак, мы переходим к *изменяемым* типам объектов. Основных их - три: *списки*, *множества* и *словари*. Сейчас мы будем заниматься списками и множествами, а словари оставим на потом. Также мы вкратце рассмотрим неизменяемый тип *кортеж* (tuple), который во всем похож на список за исключением того, что он не изменяется.

Все объекты перечисленных типов являются *структурами данных*. Это означает, что каждый объект такого типа содержит в себе другие объекты, определенным образом организованные. Вся аналитика данных является работой с большими структурами данных, преобразованием одних структур в другие, оптимизацией поиска, хранения и передачи таких структур.

Также списки, кортежи, множества и словари называют *коллекциями*. Это название используется чаще, чем словосочетание "структуры данных", но оно менее точно, поскольку акцентирует только то, что объект содержит другие объекты, но не его внутреннюю организацию. Строка также является коллекцией, потому что состоит из отдельных символов. Существуют в питоне и другие типы коллекций, полный список вы можете посмотреть в документации библиотеки collections.

#### Списки и их методы

Важный и постоянно используемый тип данных - это список (list). Список выглядит как набор каких-то элементов в квадратных скобках:

    [1, 2, 3] - это список int.
    ['1', '2', '3'] - это список str.
    [True, False, False] - это список bool.
    [[1], [2], [3]] - это список списков, каждый из которых содержит один int

Элементы могут быть не только одного, но и разных типов (смешанные списки).

Список - это итерируемый изменяемый объект. Все элементы в списке индексируются, точно так же, как в строке (следовательно, к спискам тоже применимы срезы).

Как можно создать новый список?

1. Прямо записать его (не забывайте, что объекты разных типов пишутся в питоне строго так, как это принято для конкретного типа: так питон понимает, что нам нужно число или строка):

        A = [1, 2, 3]

2. Завести пустой и добавлять в него элементы по одному с помощью метода append(). Пустой список создается либо просто квадратными скобочками, либо явно функцией list().

        A = []
        В = list()
        A.append(5) # A = [5]
        В.append('пять') # B = ['пять']

3. Преобразовать другой итерируемый объект:

        A = list('qwerty') #создает список из отдельных символов строки 'qwerty'
        B = list(input().split()) #создает список из слов введенной строки, разделяя их по пробелам

4. Заранее проинициализировать список дефолтными значениями (нулями или пустыми строками) и затем заполнять его. К такому списку можно обращаться по индексам, в отличие от пустого списка, где A[i] вызовет ошибку Index out of range.

```python
A = [0] * 5 #для списка длиной 5 элементов
A[2] = 1
A[4] = 2
print(A)
```

Какие операции можно выполнять со списками?

Как мы знаем, все арифметические и логические операторы применяются к числам; сложение, умножение и логические операторы также применяются к строкам (две строки можно конкатенировать в одну, а также одну строку умножить на int).

Списки тоже можно складывать и умножать на число. Эффект такой же, как у строк:

    [1, 2, 3] + [4, 5, 6] = [1, 2, 3, 4, 5, 6]
    [1, 2, 3] * 2 = [1, 2, 3, 1, 2, 3]

Обратите внимание, что, поскольку все элементы списка хранятся под уникальными номерами-индексами, в самом списке а) могут быть повторяющиеся элементы) б) порядок элементов строго определен.

Наконец к элементам списка можно применять все те штуки, которые применяются к типу данных, к которому относится элемент.

```python
A = [""] * 2
A[0] = 'qwerty'
A[0] = A[0].replace('q', 's')
A[1] = 'asdfgh'
print(A[0] + A[1])
```

**Методы списков**

1. Как вы уже видели, добавлять элементы в список можно с помощью метода list.append(elem): такой элемент всегда добавляется в конец списка. Заметьте, что, если в аргумент метода передать список, этот список станет элементом, а не добавится к исходному "на равных":

        A = [1, 2]
        B = [3, 4]
        A.append(B) = [1, 2, [3, 4]]

2. Есть метод, который работает так же, как +:

        A.extend(B)

3. Можно посчитать количество вхождений какого-то элемента в списке:

        A.count(elem)

4. Удалять элементы в списке можно по-разному.

        del A[i]
        A.pop(i) - не просто удалит i-ый элемент, но и вернет его (можно присвоить в переменную)
        A[i:i + 1] = [] - хитрый способ через срезы.

5. Для удаления по значению есть метод .remove(), он удаляет первое вхождение елемента в список. Если такого элемента в списке нет, получим ValueError.

        A.remove(elem)

6. Вставлять элементы тоже можно двумя способами:

        A.insert(index, elem)
        A[i:i] = [elem]

Будьте внимательны! Если вы используете итерацию по списку, то после удаления или вставки элемента индексы сдвигаются.


```python
A = [0, 1, 2, 3]
for elem in A: #сначала в списке 4 элемента, и мы перебираем их по порядку
    if elem <2: #условию удовлетворяют два первых элемента
        A.pop(elem) #удаляем по значению
print(A) #оказывается, что мы удалили только первый элемент, а второй пропустили из-за сдвига индексов
```

Еще есть два метода строк, связывающие их со списками: .split() и .join(). Метод .split() разрезает строку на несколько строк по указанному разделителю и помещает эти строки в список. По умолчанию разделителем считается пробел, но можно указать и другой, например, символ подчеркивания.

```python
string = "мама_мыла_раму"
A = string.split("_")
print(A)
```

Метод .join() превращает список в строку, объединяя элементы списка через указанный символ. Обратите внимание, что элементами списка могут быть не только строки, но и числа, которые питон превращает в строки перед объединением.

```python
A = ['4', 'черненьких', 'чумазеньких', 'чертенка']
print(" ".join(A))
```

К спискам также применяются следующие функции (не методы!):
- len(list) - возвращает длину списка (сколько в нем элементов)
- min(list) - возвращает самый маленький элемент
- max(list) - возвращает самый большой элемент
- sum(list) - возвращает сумму всех элементов, если все элементы - числа.

Как сравниваются между собой списки? Так же, как строки: поэлементно.

    [1, 2, 3] < [3, 4, 5]
    [1, 2, 3] < [1, 3, 4]
    [1, 1, 3] < [1, 1, 4]
    [1, 1, 1] == [1, 1, 1]

То есть, питон сперва сравнивает первые два элемента между собой, если они равны, то сравнивает следующие два элемента. Как только он доходит до момента, когда парные элементы не равны, то приписывает < или >.

**Копирование списков**

Напомню, что питон - язык, который берет на себя управление памятью за вас. С одной стороны, это очень здорово, потому что вам не приходится думать о том, как положить данные в оперативу; с другой стороны, питон считает себя умнее вас и поэтому специфично ведет себя с изменяемыми и неизменяемыми объектами, за что его не любят многие программисты С++ и других языков.

О чем это?

Когда у нас есть неизменяемый объект, например, число, и мы хотим записать его в другую переменную:

    x = 5
    y = x
    x += 1

y остается все равно 5, потому что в момент, когда мы сделали x = 6, переменная x просто стала указывать на новый объект (помните, что числа не изменяются, а просто теряются в памяти?)

Но когда у нас объект изменяемый, то операция присвоение B = A просто создаст еще одну ссылку на тот же самый объект, а не скопирует его.

```python
A = [1, 2, 3]
B = A
B[0] = 4
print(A)
```

То есть, объект "список" у нас один: но на него указывают две переменные. И если мы поменяем часть объекта по переменной В, то это отразится и в переменной А тоже!

Так сделано для того, чтобы глупенький программист не копировал списки (а они обычно огромные и занимают много памяти) зря. А нам, глупеньким программистам, приходится постоянно про это помнить, и случай в ячейке выше - еще самый простой!

Явно откопировать один список в другой можно при помощи метода copy:

    B = A.copy()

Также можно использовать для этой цели пустые срезы:

    B = A[:]

Но в общем и целом питон обычно прав, и не копируйте списки без особой нужды!

**Сортировка и переворачивание списков**

Списки можно сортировать двумя способами.

1. С помощью метода .sort()

        A = [2, 5, 3, 4]
        A.sort()

Метод изменяет список *на месте*: если вы напишете B = A.sort(), в переменную B положится None (пустота).

2. С помощью функции sorted()

        B = sorted(A)

Функция sorted() принимает на сортировку не только списки, но и любые итерируемые объекты, и возвращает их в виде списка. Поэтому результат работы функции можно и обычно нужно присваивать в переменную, а еще можно ей передавать строчки.

И у метода, и у функции есть один параметр: key, который умеет принимать функцию, по результатам которой будет сортировать.

Например, если мы хотим отсортировать список строчек не по алфавиту, а по длине строки:

```python
A = 'Richard of York gave battle in vain'.split()
print(A)
B = sorted(A, key=len)
print('B:', B)
A.sort(key=len, reverse=True)
print('A:', A)
```

В методе .sort() я здесь сразу использовала еще один параметр: reverse, который по умолчанию считается False, но если его явно указать как True, сами посмотрите, что будет.

Обратите внимание, что в параметре key функция передается без скобочек. Потому что здесь она не *вызывавется*, мы только сообщаем функции sorted(или методу sort), какую функцию нужно использовать. Функция вызывается уже где-то там в недрах сортировки. Конечно, мы должны использовать такие функции, которые умеют обрабатывать наш сортируемый тип данных и при этом что-нибудь возвращают! (Иначе зачем все это?..)

Аналогичным образом перевернуть список задом наперед можно двумя способами:

1. Методом reverse


    A.reverse()


2. Функцией reversed:


    B = reversed(A)

В случае с функцией нас, правда, ожидает маленькая засада, потому что функция возвращает не список, а какой-то другой, непонятный нам объект - *итератор*.

```python
A = [4, 6, 2, 7, 3, 9]
A.reverse()
print('New A:', A)
B = reversed(A)
print(B)
```

Однако этот объект можно легко превратить в список с помощью функции list().

#### Кортеж (tuple)

Кортежам обычно в курсе программирования почти не уделяется внимания, потому что по функционалу они близки к спискам, единственное их отличие в том, что их нельзя изменять. Кортеж - это набор каких-то элементов; обычно они указываются в круглых скобочках, но можно задавать кортеж и без скобочек.

    T = (1, 2, 3)
    T = 1, 2, 3

Это создаст одинаковый кортеж в обоих случаях.

Из-за своей неизменности кортежи обрабатываются быстрее списков и удобны тогда, когда нам нужен некий набор констант.

У кортежей есть только два встроенных метода - это .count() для подсчета количества вхождений определенного элемента и .index() для поиска первого вхождения элемента. Они работают в точности так же, как одноименные методы списков.

В связи с кортежами следует упомянуть такую возможность питона, как *множественное присваивание*.

Внимательно считайте, сколько объектов стоит слева и справа!

    x, y = 4, 5 # присвоит 4 в x, а 5 в y
    x = 4, 5 # сделает в x кортеж (4, 5)
    x, y = 4 # вызовет ошибку!

Умный питон интерпретирует код справа от оператора присваивания по-разному в зависимости от того, сколько переменных мы ему указали слева.

Множественное присваивание можно использовать и со списками:

        x, y = [1, 2] # присвоит 1 в x, а 2 в y

С помощью множественного присваивания можно легко менять переменные местами, как показано ниже.

```python
x = 4
y = 5
x, y = y, x
print(x, y)
```

Питон сперва вычислит, что у него лежит справа (4, 5), а потом присвоит это в левую сторону. Так ничего не потеряется!

**Создание списков пар. Функции enumerate() и zip()**

Часто используемый в питоне вид кортежей - это упорядоченная пара объектов, в которой первый элемент пары принадлежит к какому-то одному типу данных, а второй - к другому. Например, в таком виде можно хранить пары из имени человека и номера его телефона. (Более продвинутый способ хранения таких пар предоставляют словари, с которыми мы познакомимся позже).

При этом из самих таких пар можно образовывать другие итерируемые объекты. Существует несколько удобных способов делать это. Мы сейчас рассмотрим два из них - функции enumerate() и zip().

Функция enumerate() предназначена для того, чтобы одновременно перебирать индексы и элементы какого-то итерируемого объекта. Ее удобно использовать с множественным присваиванием.

```python
fruits = ['banana', 'apple', 'plum', 'peach']
for i, elem in enumerate(fruits):
    print (f"Position {i+1} is {elem}.")
```

Функция zip() принимает в качестве аргументов два произвольных итерируемых объекта и соединяет и попарно их элементы. При этом, если длины объектов различаются, она выдает объект меньшей длины, и соединяет элементы с начала до того момента, пока в одном из наборов они не кончатся.

```python
A = [4, 6, 2, 7, 3, 9]
B = ("е","f", "g", "h")
print(list(zip(A, B)))
```

Обратите внимание, что zip() и enumerate() возвращают не списки, а опять-таки итераторы, поэтому если мы хотим увидеть результат их работы, нам нужно сначала преобразовать его, например, в список.

Если вам интересно, что такое итераторы и зачем они нужны, почитайте про них документацию. Но если в двух словах, итератор - это объект, который хранит не сами элементы, а способ их создания, потому что так экономится память и вычислительные ресурсы. Сами элементы вычисляются по мере необходимости, например, когда мы присваиваем их в какую-то переменную. Этот подход в программировании называется *стратегия ленивых вычислений*.

#### Множества. Методы множеств

Множество - это изменяемая и итерируемая коллекция элементов. Множества работают в питоне точности как множества математической логики.

Множества отличаются от списков и кортежей, в первую очередь, тем, что они не упорядочены. Поэтому во множестве могут находиться только хешируемые (неизменяемые) объекты: питону же надо как-то быстро их находить, а если индексов нет, он ищет объекты по хешу. То есть если мы попытаемся положить в множество кортеж, все будет хорошо, но если список, вызовется ошибка.

С этим же связано то, что в множестве могут быть только *уникальные* объекты. Если мы попытаемся добавить во множество объект, который там уже есть, ничего не произойдет.

Как задавать множества?

1. Явно в коде:

        S = {1, 2, 3}

2. Завести пустое множество и добавлять в него элементы.

        S = set()
        S.add(5)

Обратите внимание, что множества, в отличие от списков, нельзя задавать просто {}: так получится не множество, а словарь! Всегда пишем только set().

3. Явное преобразование:

        A = [1, 2, 3]
        S = set(A)

**Методы множеств**

1. Добавляет элемент в множество: A.add(elem)

2. Удаляет конкретный элемент: A.discard(elem)

Обратите внимание, что при отсутствии указанного элемента метод .discard() ничего не будет делать.

*Операторы множеств и соответствующие им методы*

Главное отличие операторов от методов в том, что у оператора и слева, и справа обязательно должно стоять множество, а метод может в скобочках принимать любой итерируемый объект.

1. Конъюнкция (объединение):

        A | B
        A.union(B)

2. Дизъюнкция (пересечение):

        A & B
        A.intersection(B)

3. Разность:

        A - B
        A.difference(B)

4. Симметрическая разность:

        A ^ B
        A.symmetric_difference(B)



```python
A = {1, 2, 3}
B = {2, 3, 4}
print('Union:', A | B)
print('Intersection:', A & B)
print('Difference A - B:', A - B)
print('Difference B - A:', B - A)
print('Symmetric difference:', A ^ B)
```

Интересно, что любые логические выражения могут рассматриваться как множества. Соответственно, методы множеств могут рассматриваться как эквиваленты логических связок. Вы можете даже использовать их в коде вместо соответствующих словесных обозначений (но лучше без нужды так не делать, потому что это ухудшает читаемость).

- \| - это обычная связка "или", т.е. A | B может использоваться вместо A or B,

- \& - это связка "и", т.е. A & B может использоваться вместо A and B,

- ^ - это исключающее "или", для которого в питоне нет специально зарезервированного слова, она может быть выражена словесно только с помощью составной конструкции (A or B) and not (A and B)

В том, что такие замены работают корректно, вы можете убедиться ниже.

```python
A = int(input("Введите логическое значение А (1 или 0): "))
B = int(input("Введите логическое значение B (1 или 0): "))
print("A | B", bool(A) | bool(B))
print("A & B", bool(A) & bool(B))
print("A ^ B", bool(A) ^ bool(B))
```

**Проверки множеств**

Это про понятие подмножества и надмножества. Здесь питон предоставляет нам только методы, специальных значков для отношения включения, как в математической логике, в нем нет.

```python
A = {1, 2, 3, 4, 5}
B = {1, 2, 3}
C = {6, 7, 8}
print('А является надмножеством В:', A.issuperset(B))
print('В является подмножеством А:', B.issubset(A))
print('А и С не пересекаются:', A.isdisjoint(C))
```

**Финальные замечания**

К множествам применимы некоторые методы, работающие на списках. Например, есть такой же метод удаления элемента по значению .remove(). В отличие от метода .discard(), метод remove() вызывает ошибку, если указанный элемент не найден.

Также для множеств есть специальное применение метода .pop() без аргумента, в результате из множества удаляется один элемент, который мы получаем на выходе, но заранее не известно, какой.

```python
A = {"b", "с", 4, 7}
a = A.pop()
print(A, a)
```
