####Задача 6.

Разберитесь, как работает функция склонения русских прилагательных из кода ниже. Расширьте ее как-нибудь, например, добавьте поддержку прилагательных с основой на -ц типа "куцый". Добавьте соответствующие кейсы в тест, чтобы проверить, что все работает правильно.

```python
def decline_adjective(masculine_form, gender, number="singular"):
    """
    Правильное склонение русских прилагательных.
    """
    if masculine_form.endswith("ый") or masculine_form.endswith("ой"):
        # Твёрдое склонение
        if masculine_form.endswith("ой"):
            stem = masculine_form[:-2]
        else:  # "ый"
            stem = masculine_form[:-2]

        if number == "singular":
            if gender == "m":
                return masculine_form
            elif gender == "f":
                return stem + "ая"
            elif gender == "n":
                return stem + "ое"
        else:  # plural
            # После шипящих пишем "ие", иначе "ые"
            if stem.endswith(("ж", "ш", "ч", "щ")):
                return stem + "ие"
            else:
                return stem + "ые"

    elif masculine_form.endswith("ий"):
        stem = masculine_form[:-2]
        last_char = stem[-1] if stem else ""

        # Определяем тип основы
        if last_char in "жшчщ":
            # Шипящие: хороший
            if number == "singular":
                if gender == "m":
                    return masculine_form
                elif gender == "f":
                    return stem + "ая"
                elif gender == "n":
                    return stem + "ее"
            else:  # plural
                return stem + "ие"

        elif last_char in "хгк":
            # Заднеязычные и х: тихий, строгий
            if number == "singular":
                if gender == "m":
                    return masculine_form
                elif gender == "f":
                    return stem + "ая"
                elif gender == "n":
                    return stem + "ое"
            else:  # plural
                return stem + "ие"

        else:
            # Мягкое склонение: синий
            if number == "singular":
                if gender == "m":
                    return masculine_form
                elif gender == "f":
                    return stem + "яя"
                elif gender == "n":
                    return stem + "ее"
            else:  # plural
                return stem + "ие"

    return masculine_form


# Тестирование
test_cases = [
    # Твёрдое склонение
    ("новый", "f", "singular", "новая"),
    ("новый", "n", "singular", "новое"),
    ("новый", "f", "plural", "новые"),
    ("большой", "f", "singular", "большая"),
    ("большой", "n", "singular", "большое"),
    ("большой", "f", "plural", "большие"),

    # Мягкое склонение (синий тип)
    ("синий", "f", "singular", "синяя"),
    ("синий", "n", "singular", "синее"),
    ("синий", "f", "plural", "синие"),

    # Шипящие (хороший тип)
    ("хороший", "f", "singular", "хорошая"),
    ("хороший", "n", "singular", "хорошее"),
    ("хороший", "f", "plural", "хорошие"),

    # Заднеязычные и х (тихий тип)
    ("тихий", "f", "singular", "тихая"),
    ("тихий", "n", "singular", "тихое"),
    ("тихий", "f", "plural", "тихие"),
    ("строгий", "f", "singular", "строгая"),
    ("строгий", "n", "singular", "строгое"),
    ("строгий", "f", "plural", "строгие"),
]

for adj, gender, number, expected in test_cases:
    result = decline_adjective(adj, gender, number)
    status = "✓" if result == expected else f"✗ (получено {result})"
    print(f"{adj} → {result} (должно быть {expected}) {status}")
```
