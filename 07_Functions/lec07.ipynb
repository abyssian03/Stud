#### Идея и синтаксис функции

Снова немного вернемся к истории и парадигмам программирования.

Итак, для маленьких коротких программ парадигма структурного программирования была достаточной, чтобы сделать их ясными и читаемыми, но программы все росли, и сегодняшние проекты могут достигать многих тысяч строк кода. Тут уже простая структура не спасет! Нужно прибегать к более сложным методам организации кода.

Для этого в современном мире программирования обычно различают два противоположных подхода:

- функциональное программирование
- объектно-ориентированное программирование

В каком-то смысле это соотносится с категориями "глагол" и "существительное" в естественном языке.

И тот, и другой взгляд на вещи помогают объединять части кода в какие-то категории, но по разным признакам.

Функциональное программирование - это "глагольный" подход: мир состоит из списка действий, которые можно выполнять, а субъекты и объекты не важны. Так, в нашем мире есть функция "приготовить обед", а кто будет в ней участвовать - неважно, может и котик обед готовить, разница будет только в результате.

Объектно-ориентированное программирование - это "существительный" подход: мир состоит из объектов, у которых есть разные атрибуты и характеристики. Объект "котик" умеет мурлыкать и мяукать, это его характеристики.

Функция - это такой объект, который состоит из заранее собранного кода. То есть, допустим, если нам нужно какой-то набор действий выполнять несколько раз в программе, мы можем этот набор, так сказать, вынести за скобки, сохранить под отдельным и вызывать каждый раз, как только нам это понадобится.

Функция - вызываемый объект.

(Питон первоначально задумывался как объектно-ориентированный ЯП, поэтому, хотя он позволяет нам использовать функциональную парадигму, у него все равно все внутри представляется как объекты).

У функции есть две фазы существования: место, где она определяется в программе, и место, где она вызывается.

Определять функции можно где угодно, но принято делать это в самом начале по конвенциям.

Шаблон определения функции:

    def funcname(*args, **kwargs):
        <тело функции>
        return <some values>

Что здесь что?

def - это слово, сообщающее питону, что мы хотим определить функцию.

funcname - имя нашей функции, которое мы придумываем сами. Принято писать имена функций строчными буквами. Это такая же переменная!

*args - означает неопределенное количество позиционных аргументов, то есть таких, которые подаются без имени, и питон просто по позиции в ряду понимает, какому аргументу присвоить какое значение.

**kwargs - означает неопределенное количество именованных аргументов, то есть таких, которые переданы по ключу (штуки наподобие sep и end у print - это именно аргументы, переданные по ключу).

В теле функции могут быть любые команды, ветвления, циклы.

return - оператор возвращения; его может не быть, тогда функция возвращает None. Если написать просто return без всего, тоже вернется None. Если в return прописать какой-то объект, то наша функция будет его возвращать, и его можно будет складывать в переменную или как-то иначе с ним потом работать.

```python
def area(r):
  """Функция, которая возвращает площадь круга по его радиусу"""
  return 3.14 * (r ** 2)

r = int(input('Введите r: '))
s = area(r)
print('Площадь равна', s)
```

#### Виды аргументов

Мы уже несколько раз упоминали аргументы и параметры. На самом деле, если подходить к терминологии строго, то параметры - это то, что мы пишем в скобках функции во время ее определения, а аргументы - это конкретные объекты, которые мы ей передаем во время вызова, но обычно все это забывают, включая разрабов.

Параметры в функции можно задавать как
- обязательные
- необязательные

Обязательные параметры необходимо будет передать при вызове функции, а необязательные имеют какие-то дефолтные значения.

```python
def myfunc(arg1, arg2=4):
  print(arg1, arg2)

myfunc(10, 5)
myfunc(10)
```

Эти разновидности относятся только к определению функции!

С другой стороны, передавать аргументы в функцию во время **вызова** можно тоже двумя способами:

- позиционно
- по ключу (keyword)

**Главное правило**: позиционные аргументы всегда должны идти до именованных.



```python
def myfunc(arg1, arg2, arg3=3):
  print(arg1, arg2, arg3)

myfunc(1, 2, 3)
myfunc(arg2=1, arg1=2)
myfunc(arg3=10, arg2=20, arg1=25)
```

Обратите внимание: то, как мы задаем параметры в определении функции, и то, как мы их передаем в вызове функции - взаимосвязанные, но не зависящие друг от друга напрямую вещи! Вы можете указывать имя параметра, даже если он определен как позиционный. Это улучшает читаемость кода, позволяя не соблюдать строгий порядок аргументов.

**Исключение**: Использование символа / в определении функции делает аргументы перед ним строго позиционными, их нельзя передать по имени.



#### Аннотирование типов

Все вышеописанное касается того, как питон работает с функциями; но есть еще способы подробнее описать свою функцию: ее работоспособность это никак не изменит, а IDE начнет подсказывать по вашим описаниям.

```python
def myfunc(arg1: list, arg2: int=2) -> str:
  return str(arg2)

myfunc([1, 2, 3], 3)
```

Питон - язык с неявной типизацией, но с версии Python 3.5+ можно использовать подсказки типов, которые проверяются линтерами, но игнорируются интерпретатором. Аннотирование типов помогает улучшить читаемость и позволяет использовать статический анализ, который проверяет код на наличие ошибок несоответствия типов без его запуска.


Функция в ячейке выше ничего осмысленного не делает. Когда вы начнете набирать название своей функции в теле программы, чтобы ее вызвать, питон подскажет, что он ожидает, что аргумент arg1 должен быть списком, а arg2 - числом. -> str сообщает и нам, и питону, что функция должна возвращать str (хотя не обязано), и он будет подсказывать нам, если мы попытаемся применить какой-нибудь не строчный метод к объекту функции, что это не должно работать корректно. А в IDE, если вы укажете -> str и напишете при этом return 0, среда разработки может выделить ваше 0 и написать, что ожидается str.

Также можно использовать оператор pass, который ничего не делает и служит филлером: если хотите написать функцию, но тело функции еще не придумали, можно написать pass и дописать потом.

#### Пространства имен и области видимости

Самое сложное, что связано с функциями - это понятие области видимости переменных.

В питоне различаются глобальные и локальные переменные. Обычные переменные в теле программы - глобальные, они существуют во все время работы программы. Переменные в теле функции - локальные и существуют только во время работы функции.

Как это происходит? К каждой функции привязывается так называемое **пространство имен**, которое можно понимать как словарь, где ключами являются имена переменных, а значениями - объекты, точнее, их адреса в памяти. В случае вложенных функций пространство имен внутренней функции также является вложенным по отнощению к пространству имен внешней функции, которое в этом случае называется объемлющим (enclosing). Все пространства имен функций верхнего уровня вкладываются в глобальное пространство имен файла (**модуля**), а само оно вкладывается во встроенное пространство имен самого питона, которое содержит имена встроенных функций - таких как print, - f nfr;t имена и коды ошибок. Встроенное и глобальное пространства имен создаются при интерпретации кода в файле, а локальные пространства имен - при вызове функций.

Области видимости переменных организованы так, что из функций можно можно обратиться к переменным глобального и объемлющего пространства имен, но наоборот сделать нельзя: локальные переменные видны только из функции, пространству имен которой они принадлежат.

    def func(arg):
      arg = 3
      print(arg)
      arg *= 8
      print(arg)

    x = 5
    func(x)
    print(x)
    print(arg)

В такой бессмысленной программе переменная x будет существовать все время и в любой момент с ней можно что-то делать; переменная arg будет существовать только в момент, когда мы делаем func(x) - при этом питон неявно выполняет arg = x - и вызов print(arg) заставит питон заругаться, что он не знает такой переменной.

Таким образом, всякий раз, когда вы обращаетесь к переменной по имени изнутри какой-то функции, питон ищет это имя сначала в локальном пространстве имен, затем в объемлющем, в случае вложенных функций, и только потом в глобальном. То же самое происходит с именами самих функций.

Получив доступ к глобальной переменной из локальной области видимости, мы можем прочитать ее значение, но не можем просто так изменить его. Это происходит потому, что в момент, когда мы обращаемся к глобальной переменной из функции и пытаемся что-то с ней сделать, у нас появляется локальная переменная с тем же именем, которая "затеняет" глобальную, а глобальная на время становится недоступной. Мы работаем с локальной переменной, меняем ее значение, но глобальная переменная остается незатронутой.

Когда функция отрабатывает, глобальная снова возвращается на сцену, потому что локальная перестала существовать.

```python
x = 10

def myfunc():
    print(x)
    print('Local x is', x)

my_function()
print('Global x is', x)
```

Но что если мы хотим изнутри функции поменять глобальную переменную так, чтобы при завершении функции это значение таким и осталось, не сбросилось? Для этого есть специальное средство - объявление переменной в начале функции как глобальной. Оно производится с помощью управляющего слова global. После такого объявления все изменения, которые произошли со значением этой переменной внутри функции, сохранятся после окончания ее работы.

```python
counter = 0

def increment():
    global counter  # Указываем, что работаем с глобальной переменной
    counter += 1
    print(counter)

increment()  # Выведет 1
```

#### Лямбда-функции

Допустим, у нас есть функция-крошка:

```python
def microfunc(x):
  return x ** 3
```

Эта функция ничего не делает, а только возвращает куб от своего аргумента. Сколько мишуры! Нужно написать def, придумать название функции, поставить двоеточие, написать оператор return.

В питоне для таких микрофункций есть упрощенный синтаксис, который позволяет задавать их в одну строку.

```python
microfunc = lambda x: x ** 3
```

Управляющее слово lambda говорит питону, что мы сейчас создадим объект типа "функция", а название ему придумывать не хотим (я в ячейке выше присвоила объект в переменную microfunc, и теперь этот объект можно вызывать как microfunc(), но это было совсем не обязательно).

Лямбдой можно задать только такую функцию, у которой в теле нет ничего, кроме return. Эти функции удобно использовать в ключе сортировки и во многих других местах. Можно их задавать так, как я сделала в ячейке выше, это просто укоротит запись.

```python
A = [123, 456, 33, 1234]
A.sort(key=lambda x: x % 10)
print(*A)
```

Такая сортировка позволяет нам отсортировать числа не обычным способом, а по последней цифре (потому что x % 10 и возвращает последнюю цифру многозначного числа).

У лямбда-функции может быть и несколько аргументов, в этом случае они перечисляются через запятую:

    lambda x, y, z: ...

Аргументы могут быть любых типов, в том числе они сами могут быть функциями. Но чаще происходит наоборот: лямбда-функции выступают в качестве аргументов обычных функций, потому что их можно задать в одну строчку. Особенно это удобно, когда при указании в функции в качестве ее именованного аргумента другой функции мы указываем значение по умолчанию.

```python
def operation_ex(x, operation=lambda x: x**2):
    return operation(x)

print(operation_ex(2))
print(operation_ex(2, lambda x: x**3))
```

В лямбда-функциях могут использоваться условия и генераторы, их синтаксис в этом аспекте не представляет собой ничего нового. Вот пример лямбда-функции с условием:

    maxx = lambda x, y: x if x > y else y
