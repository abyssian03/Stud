### Арифметические операции и операторы сравнения

В принципе все математические штуки воплотимы в питоне: арифметические операции, операции сравнения, логические операции и операции на множествах. С множествами мы будем работать позже, а сейчас разберем первые три пункта.

Что касается арифметики, нас есть стандартный набор операторов:

+ - омоним оператора конкатенации на строках, с числами работает как обычное сложение;

- - тоже вполне обычное вычитание;

* - помним, что; помимо умножения числа на число, доступно также умножение строки на число;

/ - обычное деление в питоне обязательно неявно преобразует целые числа к float, то есть оно вернет дробь, даже если делится нацело;

// - целочисленное деление;

% - остаток от деления;

** - степень -- обратите внимание, что она пишется не символом ^, как она иногда обозначается в математике, а только так!

Скобки работают как и в школьной математике: сначала выполняется операция в скобках, а потом ее результат используется для вычисления выражения за скобками. Без скобок порядок применения операций тоже такой, как в школьной математике: точно так же возведение в степень выполняется в первую очередь, за ним умножение и деление, а сложение и вычитание в конце.

К любому из этих операторов можно приписать =, тогда оператор применится к старому значению переменной и новому числу, а результат запишет в опять в ту же переменную. Это дает эффект сокращения записи. Например, выражения

x = x + 1

и

x += 1

будут действовать одинаково - увеличивать значение переменной x на единицу.

```python
# простой алгоритм выделения цифр из трехзначного числа с помощью арифметических операций целочисленного деления и остатка
x = int(input("Введите трехзначное число: "))
units = x%10
x //= 10
dec = x%10
hun = x//10
print(hun, dec, units)
```

Также у нас есть все обычные операции сравнения:

\> - больше,

< - меньше,

\>= - больше или равно,

<= - меньше или равно,

== - равно,

!= - не равно.

Две последних операции работают не только с числами, но и с другими типами данных, например, они могут сравнивать значения двух строковых переменных.

Очень важно не путать = с ==. Первое - оператор присваивания, второе - оператор сравнения.

Возьмите на заметку: результатом действия операторов сравнения на целые или дробные числа являются не такие же числа, а всегда объекты типа bool. Этим операторы сравнения отличаются от арифметических операций.

В целом, практически любой объект можно превратить в bool. Помним, что, если объект - число, то все, кроме нуля, превратится в True, а ноль - в False. Если объект - строка, то пустая строка превратится в False, непустая - в True. Точно так же, как со строками, bool() работает со значениями типа список, словарь или множество (мы увидим это позже).

```python
bool('False') # как превратить ложь в истину?
```

### Логические операторы (коннективы)

Логические значения можно соединять и преобразовывать с помощью так называемых **коннективов** - логических операторов

and - логическое "и",

or - логическое "или",

not - логическое отрицание.

Скобки опять же работают, как и везде!

```python
x = 5
print((x > 0 and x % 2 != 0) or (x < 0 and x % 5 == 0))
```

Обратите внимание, что логические "или" возвращает True в любом случае из трех: если выполнена первая часть условия, если выполнена вторая или если выполнены обе части. Исключающее "или" тоже есть в питоне, оно обозначается символом ^ (теперь вы понимаете, почему нужно помнить, что степень числа обозначается иначе!) и возвращает True только в том случае, если выполняется какая-то одна из частей условия, но не обе вместе.

Порядок применения условных операторов такой же, как в логике: первым применяется not, затем and, а в конце or. Однако для удобства читаемости кода части сложных условия полезно выделять скобками, даже если порядок и без них был бы правильным (как в примере выше).

### Условные инструкции

Очень часто в программировании есть необходимость произвести действие, только если выполняется некое условие, а при нарушении условия не производить никакого действия или произвести другое действие. В таких случаях используется так называемое **ветвление**, которое в питоне оформляется с помощью условной конструкции

    if <condition>:
        command
        ...
    elif <condition>:
        command
        ...
    elif <condition>:
        command
        ...
    ...
    else:
        command

Это общий шаблон условной конструкции, но она может быть и короче. Ветки elif и/или else могут отсутствовать, обязателен только if, после которого должна быть хотя бы одна команда.

Ветки if и elif могут сожержать в качестве условия какое-нибудь выражение (a == b) или объект ('abc'); выражение при этом вычисляется и дает в качестве результата bool, а объект неявно преобразовывается в bool по описанным выше правилам.

Поэтому грамотная и часто используемая проверка:

    if a:
        <do something>

Обратите внимание, что ветка else не содержит никакого условия и выполняется во всех случаях, когда не выполняется ни одно из условий выше.

Также обратите внимание, что условия ветки elif проверяются только в тех случаях, когда условие ветки if не выполнено. Если программа прошла по ветке if, остальная часть условной конструкции игнорируется.

```python
x = int(input("Введите целое число: "))
if x // 2 == 0:
    print("Число четное")
else:
    print("Число нечетное")
```

Обратите внимание на синтаксис условной конструкции: после if <condition> ставится двоеточие, и следующий блок кода пишется с отступом в четыре пробела (в большинство IDE можно один раз нажать Tab, тогда автоматически выставится отступ нужной ширины). Двоеточие в конце строки и отступ в начале следующей представляют собой парные элементы синтаксиса: если интерпретатор встречает двоеточие, он автоматически ожидает найти далее отступ.

Соблюдать правильные отступы очень важно, поскольку это единственное средство обозначения подчиненных блоков кода в питоне (в С для этого применяются фигурные скобочки). Если где-то вы вставили один лишний пробел или, наоборот, не хватает одного пробела, интерпретатор на этом месте будет ругаться. Технически есть возможность писать код с отступами другой ширины, например, в два пробела, а не в четыре, но это настоятельно не рекомендуется во избежание несовместимости вашего кода с кодом других программистов в будущем. Лучше сразу привыкать к культуре, принятой в программистском сообществе, чтобы потом не пришлось переучиваться.

В случаях, когда имеются вложенные условия, отступ внутреннего условия добавляется к отступу внешнего условия, и получается ступенчатый код, который зрительно легко воспринимать.

Код ниже представляет собой пример использования вложенных условий. Разберите его внимательно.

```python
# игра "Угадай число за минимальное количество вопросов"
print("Загадайте число от 1 до 9")
answer1 = input("Ваше число четное? (Да/Нет) ")
answer2 = input("Ваше число делится на три? (Да/Нет) ")
answer3 = input("Ваше число больше пяти? (Да/Нет) ")
if answer2 == "Да":
    if answer1 == "Да":
        print("Ваше число 6")
    elif answer3 == "Да" :
        print("Ваше число 9")
    else:
        print("Ваше число 3")
elif answer3 == "Да":
    if answer1 == "Да":
        print("Ваше число 8")
    elif answer1 == "Нет":
            print("Ваше число 7")
else:
    answer4 = input("Ваше число простое? (Да/Нет) ")
    if answer1 == "Нет" and answer4 == "Нет":
        print("Ваше число 1")
    elif answer1 == "Да" and answer4 == "Нет":
        print("Ваше число 4")
    elif answer1 == "Нет" and answer4 == "Да":
        print("Ваше число 5")
    else:
        print("Ваше число 2")
```
