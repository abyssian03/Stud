### Структурное программирование. Циклы

В прошлом веке программы были еще довольно маленькие, и в них легко было разобраться. Но когда появились первые ЯП относительно высокого уровня, люди придумали команду go to, которая отправляла компьютер выполнять любую строчку, какую указывал программист, и внутренняя структура программ с этой командой становилась похожа на мешанину.

Поэтому ученый программист Эдсгер Дейкстра (https://en.wikipedia.org/wiki/Edsger_W._Dijkstra) написал целую статью про то, что давайте избавимся от этого go to и будем структурировать свой код, чтобы в нем не запутаться.

Так появилась парадигма структурного программирования. В рамках этой парадигмы предполагается, что мы, когда программируем, сперва разрабатываем строгий алгоритм, а потом пишем код в соответствии с этим алгоритмом, и никаких внезапных переходов на N-ю строчку.

В структурном программировании есть три основных понятия:
- последовательность
- ветвление
- цикл

Последовательное выполнение кода означает, что компьютер выполняет команду за командой в таком порядке, в котором они идут.

Ветвления - это условные инструкции, позволяющие выполнять один код при выполнении одних условий, и другой - при других. Мы с ними уже знакомы.

Циклы - то, что мы рассмотрим сейчас. Цикл - это блок инструкций в программе, которые повторяются или заданное количество раз, пока выполняется условие.

Таким образом, у нас есть две разновидности циклов:

1. **Определенный цикл** - когда мы заранее знаем, сколько раз нам нужно повториться, или перебираем конечный итерируемый объект. Например, когда нам нужно вывести таблицу умножения для всех цифр от 1 до 9.

2. **Неопределенный цикл** - когда мы должны повторять набор действий, пока выполняется какое-то условие. Например, мы ждем от пользователя, когда он введет цифру, а он вводит буквы, глупый человек. Мы никаким образом не можем предсказать, когда до него наконец дойдет, что нужно ввести цифру, поэтому цикл с заданным количеством итераций не подходит.

Определенный цикл - это цикл for, неопределенный - цикл while.

####Цикл for

В самом частом случае код такого цикла выглядит так:

      for i in range(n):
          command0
          command1
          ...

*i* - это переменная цикла, которая меняет свое значение, пока работает цикл. Вообще имя переменной может быть каким угодно, *i* используется просто по традиции. Именно из-за этой традиции объекты, которые можно перебрать поэлементно, называются **итерируемыми**, а единичный проход по циклу, соответствующий одному значению переменной цикла, называется **итерацией**.

range(n) - это результат применения функции range() к некоторому целому положительному числу *n*, который представляет собой отдельный вид итерируемого объекта, называемый **диапазоном**. На самом деле на месте range(n) в конструкции с for может стоять любой итерируемый объект - строка, список, кортеж, множество, словарь, итератор или генератор. Когда мы итерировали по строкам, спискам и множествам, мы по сути имели дело с частными случаями цикла for. Т.е. блок кода, идущий в цикле for после двоеточия (он называется **телом цикла**), выполняется столько раз, сколько значений принимает переменная цикла.

Итак, что же это за объект - диапазон? Это ряд чисел, который генерируется функцией range(). Если она используется с одним параметром *n*, то ряд начинается нулем и включает в себя последовательно все целые числа вплоть до *n*, но не включая само число *n*. Таким образом, ряд содержит ровно *n* элементов - это удобно! Именно потому, что нам удобно создавать с помощью параметра *n* ряд c *n* элементами, а программисты, как мы помним, начинают отсчет с нуля, а не с единицы, логично, что число *n* мы не должны включать в диапазон. Если же нам надо задать диапазон, который будет включать само *n*, мы должны, указывая параметр для range(), увеличить *n* на единицу, то есть написать:

        for i in range(n+1):
            ...

Полный синтаксис функции range() включает три параметра, аналогичных тем параметрам, которые мы использовали при создании срезов:

    range(start, stop, step)

Только начало, конец и шаг в случае диапазона передаются через запятую, а не через двоеточие, потому что здесь это аргументы функции. Все аргументы функции передаются через запятую, помним это.

Как и в случае срезов, мы можем задать отрицательный шаг, чтобы создать нисходящий ряд чисел. В этом случае так же конец диапазона (т.е. меньший параметр) не будет входить в полученный диапазон, тогда как начало (больший параметр) будет входить в него. Будьте внимательны, не запутайтесь!

Кроме того, мы можем задать range() только c двумя параметрами, и тогда они будут интерпретироваться как начало и конец диапазона, соответственно:

    range(start, stop)

Ниже дан пример конструкции for i in range() с тремя параметрами range().

```python
for i in range(1, 10, 2):
    print(i)
```

####Цикл while

Полный синтаксис цикла while выглядит следующим образом:

      while <condition>:
          command
          command
      else:
          command

Условие может быть каким угодно выражением, имеющим логическое значение или переводимым в него. Можно даже непосредственно указывать True, но тогда, чтобы закончить цикл, нужно применять альтернативный способ выхода (см. ниже), потому что True - это условие, которое всегда верно.

Цикл while очень похож на if с той лишь разницей, что в if инструкции при выполнении условия срабатывают один раз, а в цикле while они проходятся снова и снова, пока условие не будет нарушаться. Однако инструкция else все равно срабатывает только один раз в конце работы цикла, т.е. впервые, когда перестало выполняться условие.

На самом деле в цикле for тоже можно задать конечную инструкцию else, которая будет срабатывать один раз после выполнения всех необходимых итераций тела цикла. Однако else используется с циклом for крайне редко.

Ниже дан пример цикла while с инструкцией else.

```python
number = input()
while not number.isdigit():
    number = input('Введите число!\n')
else:
    print('Наконец-то число!')
```

Если мы передаем в качестве условия на выполнение цикла while объект другого типа, то происходит неявное преобразование типов, и этот объект превращается в bool. Как мы уже говорили, в случае чисел любое число, кроме нуля, превращается в True, а ноль - в False; в случае строк - любая строка, кроме пустой, превращается в True, а пустая - в False; в случае других итерируемых объектов это работает так же, как со строками.

Таким образом, абсолютно корректным (и демонстрирующим хороший стиль) является следующий код:

```python
x = int(input())
summa = 0
while x:
    summa += x
    x = int(input())
print(summa)
```

####Что еще нужно знать о циклах?

Наряду с условными инструкциями циклы - основные способы управления работой программы.

Циклы можно вкладывать друг в друга (как и условные инструкции). Обратите внимание, что индентинг (отступы) выравнивает команды и помогает визуально отличить, где какая часть программы. Иногда бывает несложно спутать, к чему относится else: смотрите, на одном уровне с какой инструкцией он находится. В IDE обычно проводятся вертикальные линии, чтобы удобно было это отслеживать.

    while <condition1>:
        if <condition2>:
            for i in <iterable>:
                command0  # эта команда - в теле цикла for!
        else: # этот else относится к if!
            command1
            command2
        command3 # эта команда - в теле цикла while!
    else: # этот else относится к while!
        command4


Будьте внимательны! При использовании вложенных циклов for называйте переменную цикла по-разному, иначе программа будет работать некорректно. Такая ситуация называется **коллизией переменных**. Вот пример.

```python
# ХОРОШИЙ КОД: используем разные имена
total = 0
for outer in range(1, 3):
    for inner in range(1, 3):
        total += inner
    total *= outer
    print(total) #ожидаем на первой итерации (0 + 1 + 2) * 1 = 3, на второй (3 + 1 + 2) * 2 = 12
print(total == 12) #True - работает корректно

# ПЛОХОЙ КОД: коллизия переменных
total = 0
for i in range(3):
    for i in range(3):
        total += i
    total *= i
    print(total) #из-за того, что значение i меняется во внутреннем цикле, имеем другой результат
print(total == 12) #False - работает некорректно
```

В цикле while тоже может возникнуть коллизия переменных, если переменная с одним и тем же именем используется для каких-то целей как во внешнем цикле, так и во внутреннем. Вот пример.

```python
i = 0
while i < 3:
    print(f"Внешний: {i}")
    i = 0  # Сброс переменной во внутреннем блоке
    while i < 2:
        print(f"  Внутренний: {i}")
        i += 1
    # Внутренний цикл завершился, i стало равно 2
    i += 1  # i станет 3, внешний цикл завершится.
```

#### break и continue
Еще у циклов есть две команды, о которых следует помнить: break и continue.

Команда break "ломает" цикл: как только программа на нее натыкается, она мгновенно выходит из цикла, не доделывая команды, которые идут за ним. Именно она позволяет закончить цикл, заданный через while true. Без этой команды такой цикл выполнялся бы бесконечно (или до тех пор, пока работа программы не будет прервана вручную), потребляя все больше и больше ресурсов, и мог бы даже заставить систему зависнуть или вылететь.

Команда continue выводит цикл из текущей итерации сразу на следующую: команды, которые за continue, не выполнятся, но цикл не остановится. Команда continue особенно удобна при работе со вложенными циклами.

Ниже даны примеры цикла while, выход из которого осуществляется только по команде break.

```python
while True:
    a = input()
    if a.isdigit():
        break
    print('Введите число!')
```

```python
while True:
    a = input()
    if a.isdigit():
        break
    if a.isspace():
        print('Нам не нужен пробел!')
        continue
    print('Вы ввели букву!')
```

Второй пример немного искусственный, однако на нем можно посмотреть, как работает команда continue.

И continue, и break могут использоваться как в цикле for, так и в цикле while, но чаще в последнем.

Нужно помнить, что если цикл снабжен блоком else, то при выходе из цикла по break этот блок не выполнится.
