## Словари, их методы и использование

####Словарь в Python

Словари отчасти похожи на множества, а отчасти на кортежи, можно сказать, что это множества кортежей, состоящих из двух элементов - внутренне упорядоченных пар, где первый элемент называется **ключом**, а второй **значением**. Чтобы подчеркнуть различные роли ключа и значения в словаре, а также для упрощения обозначений между ключом и значением, составляющим пару с этим ключом, ставится не запятая, а двоеточие, а сами пары не заключаются в скобки, просто отделяются друг от друга запятыми. Вот пример:

      dct = {key1: value1, key2: value2, key3: value3}

Сами пары "ключ: значение" в словаре не индексируются, вместо индексов для обращения к элементом словаря используются собственно ключи. Они могут быть любыми неизменяемыми объектами: числами, строками, кортежами. У ключей словаря такие же свойства, как у элементов множества: они должны быть неизменяемыми, уникальными, их порядок не определен (начиная с версии питона 3.7 гарантируется порядок по времени добавления).

Значения могут быть любыми объектами вообще. Как и список, словарь может включать в качестве значений другие словари, списки и вообще все, что угодно.

С точки зрения признаков словарь называется mapping type. По нему можно итерироваться, его можно изменять по частям. Словарь не может быть элементом множества: он изменяемый.

Как задавать словари?

1 - явно в коде:


        dct = {'sky': 'небо', 'earth': 'земля'}



2 - преобразованием из списка кортежей:


        A = [('sky', 'небо'), ('earth', 'земля')]
        dct = dict(A)



3 - преобразованием из объектов, которые возвращают функции zip() и enumerate() - они как раз возвращают списки кортежей на самом деле:


        a = ['sky', 'earth']
        b = ['небо', 'земля']
        dct = dict(zip(a, b))
        m = ['jan', 'feb', 'mar'...]
        dct = dict(enumerate(m))



4 - через генератор словарей (dict comprehension, аналогично list comprehension для списков):


        dct = {x: x ** 2 for x in range(10)}
        # соберет словарь вида 0: 0, 1: 1, 2: 4...



5 - можно собрать словарь из списка специальным методом класса:


        dict.fromkeys(list)



В последнем случае обратите внимание: мы должны написать прямо dict.fromkeys()! dict здесь не переменная, а название класса словаря.

```python
A = ['one', 'two']
dct = dict.fromkeys(A, 0) # нолик - это дефолтное значение, которое припишется всем нашим ключам.
print(dct)
```

Чтобы получить значение из словаря, можно обратиться к нему по ключу:

        print(dct['one']) - выведет значение элемента с ключом "one"

Если такого ключа в словаре нет, получим ошибку KeyError. Чтобы застраховаться от такого, можно использовать метод .get():

        print(dct['three']) - в отсутствие такого ключа выведет None

Чтобы изменить значение элемента, достаточно выполнить операцию присваивания (=):


```python
dct = {'name': 'Vasya', 'age': 23}
dct['age'] = 24
print(dct['age'])
```

Обратите внимание, что присваивание не вызывает никакой ошибки, если такого ключа еще не существует! Поэтому присваиванием можно создавать новые элементы:

        dct['gender'] = 'male'

Но если мы пытаемся совершить какую-то другую операцию с элементом словаря по несуществующему ключу, то питон выдаст KeyError. Как этого избежать? Можно либо проверять key in dct (так же, как во множестве),

        key in dct
        key not in dct

либо использовать метод .get().

И при обычном обращении по ключу, и при использовании .get() ключи можно передавать переменными точно так же, как и индексы в списке: A[i].

```python
dct = {'one': 1, 'two': 2, 'three': 3}
for key in ['one', 'two', 'three']:
    print(dct[key])
```

###view objects и способы итерации по словарю

Итак, мы выяснили, что словарь у нас - сложная структура, которая как бы состоит из двух частей: ключей и значений. В связи с этим в питоне есть особые объекты, которые называются view objects и используются в подавляющем большинстве случаев именно со словарями. Они не существуют независимо от словарей, но для каждого словаря мы можем попросить питон создать такой объект, чтобы показать нам отдельно ключи, например.

Создание view objects организовано с помощью методов словаря:

        dct.keys() - покажет все ключи словаря dct
        dct.values() - покажет все его значения
        dct.items() - покажет ключи и значения кортежами (аналогично .enumerate() для списков)

Такие view objects имеют специальные типы dict_keys, dict_values и dict_items, соответственно. Это динамические объекты, то есть они создаются не через копирование элементов словаря, а через указатели, и изменяются, когда мы вносим изменения в словарь. С этими объектами можно работать, но изменять их сами по себе нельзя.

```python
dct = {'sky': 'небо', 'earth': 'земля'}
print(dct.keys())
print(dct.values())
print(dct.items())
```

Соответственно, существует три способа итерироваться по словарю:

1 - только по его ключам. Это делается так:

        for key in dct.keys():
            ...

или просто вот так:

        for key in dct:
            ...

Поскольку чаще всего программисты хотят итерироваться по ключам, в питоне считается, что если мы написали просто dct, мы хотим ключи.

2 - только по его значениям. Это выглядит так:


        for value in dct.values():
            ...

3 - по всему сразу. Для этого пишем так:


        for key, value in dct.items():
            ...

Заметьте, что key и value - просто названия переменных, я могла назвать их i и j, просто по традиции обычно пишут key, value или k, v. В программировании много таких устоявшихся штук, как i в значении индекса, self в значении экземпляра класса и т.п. Они не обязательные, просто так удобно.

```python
monthslist = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
months = {x + 1: y for x, y in enumerate(monthslist)} #чтобы получить ключи, увеличиваем все индексы на 1
# или так:
months = dict(enumerate(monthslist, 1)) #указываем 1 непосредственно в enumerate() как параметр - начало нумерации
for key, value in months.items(): # не забываем про .items()
    print(f'{value} is number {key}')
```

####Что еще можно делать со словарями?

Со словарями умеет работать функция len(), которая вернет количество элементов в словаре.

        print(len({'sky': 'небо', 'earth': 'земля'})) - выдаст 2





Чтобы удалить элемент, можно использовать .pop():


        dct.pop(key) - удалит из словаря dct элемент с ключом key и вернет ключ
        dct.popitem() - удалит элемент, последним добавленный в словарь (если у вас питон не старше версии 3.7, иначе будете удалять случайные элементы), и вернет его в виде кортежа.

Можно использовать и del:

        del dct[key]

А метод .update() добавит один словарь в другой, при этом если у них будут одинаковые ключи, то значения в этих ключах перезапишутся:

        dct1 = ...
        dct2 = ...
        dct1.update(dct2)

```python
dct1 = {'a': 1, 'b': 2}
dct2 = {'b': 3, 'c': 4}
dct1.update(dct2)
print(dct1)
```

####Распаковка и запаковка

Есть другой удобный метод объединения двух и более словарей, свазанный с так называемым **оператором распаковки** **, который пишется перед названием словаря и превращает словарь в набор отдельных элементов:

```python
dct1 = {'a': 1, 'b': 2, 'c': 3}
dct2 = {'d': 4, 'e': 5}
dct3 = {**dct1, **dct2}
print(dct3)
```

На самом деле распаковка применима не только к словарю, но и к любому итерируемому объекту — множеству, списку, кортежу, генератору...

Но для распаковки других итерируемых объектов (не словарей) используем оператор *. Этот же оператор используется для запаковки нескольких значений в один итерируемый объект.

По сути мы уже имели дело с неявной распаковкой и запаковкой, когда изучали множественное присваивание. Потому что конструкции типа

        a, b, c = 1, 2, 3

работают именно через запаковку правой части выражения в кортеж с последующей распаковкой ее в переменные.

Синтаксис распаковки очень удобен при передаче набора значений в функцию или метод "оптом", т.е. всех сразу, сколько бы их ни было. Вот пример c функцией range().

```python
settings_lst = [(1, 10, 3), (2, 8, 2), (3, 8, 1)]
n = int(input("Выберите набор настроек (от 1-го до 3-го): "))
for i in range(*settings_lst[n-1]):
    print(i, end=" ")
```

Синтаксис запаковки используется, например, для настройки корректного множественного присваивания, когда только некоторые элементы итерируемого объекта нужно записать в отдельные переменные, а из остальных создать новый итерируемый объект. Однако помните, что, имея дело со словарем, чтобы получить значения, а не ключи, мы должны использовать метод  .values().

```python
dct = {'name': 'Vasya', 'age': 23, 'gender': 'male'}
identifier, *data = dct.values()
print(identifier, data)
```

Как видите, запаковка-распаковка представляет собой гибкий инструмент, который в зависимости от контекста может решать разные задачи. Он будет нам особенно полезен при создании пользовательских функций - мы этим займемся скоро.

#### Словари и работа с базами данных

Словари тесно связаны с таким разделом программирования как работа с базами данных. Давайте конспективно рассмотрим, в чем состоит эта связь.

- **База данных (БД)** — место, где хранятся все данные (на сервере, в файле, в облаке)
- Словарь — временное представление одной записи из БД для локального хранения в памяти и обработки программой
- Список словарей — локальная копия таблицы (или ее части) из БД

Упрощенно это можно представлять так, что база данных — это огромный шкаф с папками, а словарь — это одна папка, которую вы достали, чтобы поработать. Вот как схематично выглядит извлечение данных из базы:

Когда база данных возвращает несколько записей, мы получаем **список словарей**. Так же можно извлечь и все данные таблицы.

Мы можем фильтровать и анализировать данные прямо в Python, не делая новый запрос к БД.

```python
# Результат SQL-запроса SELECT * FROM students
students = [
    {"id": 1, "name": "Анна", "city": "Москва", "score": 85},
    {"id": 2, "name": "Борис", "city": "СПб", "score": 92},
    {"id": 3, "name": "Вера", "city": "Казань", "score": 78},
]

#Вывести всех
print("Все студенты:")
for s in students:
    print(f" {s['name']} ({s['city']}) — {s['score']} баллов")

# Найти студентов с баллом выше 80
high_score = [s for s in students if s["score"] > 80]
print("\nСтуденты с баллом > 80:")
for s in high_score:
    print(f"  {s['name']} — {s['score']}")

# Найти максимальный балл
max_score = max(s["score"] for s in students)
print(f"\nМаксимальный балл: {max_score}")

# Посчитать средний балл по группе
avg_score = sum(s["score"] for s in students) / len(students)
print(f"\nСредний балл по группе: {round(avg_score, 1)}")

# Посчитать количество студентов из Москвы
moscow_count = sum(1 for s in students if s["city"] == "Москва")
print(f"\nСтудентов из Москвы: {moscow_count}")
```

Почему это хорошо?

Во-первых, словарь организован так же, как таблица с данными. При извлечении данных в словари понятно, какие поля есть в таблице, и можно добавлять новые.

Во-вторых, работа со словарем осуществляется в программе, не влияя на состояние самой базы, мы выгружаем данные в базу только в самом конце, когда получили итоговый результат. Если программа сломалась где-то в середине, пострадают только локальные данные, а с самой базой ничего не произойдет.

В-третьих, работать с данными локально обычно быстрее, чем каждый раз обращаться к базе, даже если все, чего мы хотим - это извлечение и фильтрация данных. В реальных кейсах базы обычно огромны, и к ним одновременно могут обращаться очень много пользователей. Работа со словарями сокращает количество запросов и оптимизирует работу с базой.
